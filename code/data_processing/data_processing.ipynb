```python
#generate sensor data for vehicle 1
#change 2 places 1. vehicle_num 2. file name
import math

vehicle_num = 7

t_sensor = []
target = []
true_dis = []
noise = []
sensor_output = []

lm_all = [[0, -8, 0], [13, -3, 0], [5, 3, 0], [10, 2, 0]]
target_name_list = ['drone1', 'drone2', 'drone3', 'drone4', 'rover5', 'rover6', 'rover7', 'landmark1', 'landmark2', 'landmark3', 'landmark4']
vehicle_index = [0, 1, 2, 3, 4, 5, 6]
del vehicle_index[vehicle_num-1]

for i in range(len(pos_all[0][0])):
    v1_pos = [pos_all[vehicle_num-1][1][i], pos_all[vehicle_num-1][2][i], pos_all[vehicle_num-1][3][i]]
    new_time = pos_all[0][0][i]
    
    for j in vehicle_index: 
        t_sensor.append(np.around(new_time, decimals=3))
        new_time += 0.023
        target.append(target_name_list[j])
        dis = math.sqrt((pos_all[j][1][i]-v1_pos[0])**2 + (pos_all[j][2][i]-v1_pos[1])**2 + (pos_all[j][3][i]-v1_pos[2])**2)
        true_dis.append(dis)
        noi = np.random.normal(0,0.1,1)[0]
        noise.append(noi)
        sensor_output.append(dis + noi)
        
    for k in range(len(lm_all)):
        t_sensor.append(np.around(new_time, decimals=3))
        new_time += 0.023
        target.append(target_name_list[7+k])
        dis = math.sqrt((lm_all[k][0]-v1_pos[0])**2 + (lm_all[k][1]-v1_pos[1])**2 + (lm_all[k][2]-v1_pos[2])**2)
        true_dis.append(dis)
        noi = np.random.normal(0,0.1,1)[0]
        noise.append(noi)
        sensor_output.append(dis + noi)
        
data = {'time': t_sensor,
        'target': target,
        'true_distance': true_dis,
        'noise': noise,
        'sensor_output': sensor_output
        }

df_for_csv = pd.DataFrame(data, columns= ['time', 'target', 'true_distance', 'noise', 'sensor_output'])

df_for_csv.to_csv ('/Users/haoxuantan/Desktop/' + target_name_list[vehicle_num-1] + '_ranging.csv', header=True) #index = False

print (df_for_csv)
```
